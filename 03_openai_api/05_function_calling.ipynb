```python
from openai import OpenAI
from dotenv import load_dotenv  # .env íŒŒì¼ì˜ í™˜ê²½ë³€ìˆ˜ ë¡œë“œ
import os                       # í™˜ê²½ë³€ìˆ˜ ì ‘ê·¼ìš©

load_dotenv()                   # í˜„ì¬ ìœ„ì¹˜ì˜ .envë¥¼ ì½ì–´ì™€ í™˜ê²½ë³€ìˆ˜ë¡œ ë“±ë¡
os.environ["OPENAI_API_KEY"] = os.getenv("openai_key")  # .envì˜ openai_key ê°’ì„ OPENAI_API_KEYë¡œ ë“±ë¡
OPENWHEATHER_API_KEY = os.getenv("openweather_key")     # .envì˜ openweather_key ê°’ì„ ê°€ì ¸ì˜´
```

```python
# ë‚ ì”¨ API í˜¸ì¶œ í•¨ìˆ˜: doc_stringì„ í™œìš©í•´ í•¨ìˆ˜ ì„¤ëª…
import requests
import json

def get_current_weather(city_name='Seoul', units='metric'):
    """
    OpenWeather APIë¥¼ ì‚¬ìš©í•´ì„œ ì‚¬ìš©ìê°€ ì§€ì •í•œ ë„ì‹œì˜ í˜„ì¬ ë‚ ì”¨ ì •ë³´ë¥¼ ê°€ì ¸ì˜¤ëŠ” í•¨ìˆ˜

    Args:
        - city: str ë‚ ì”¨ì •ë³´ë¥¼ ê°€ì ¸ì˜¬ ë„ì‹œ ì´ë¦„. **ë°˜ë“œì‹œ ì˜ë¬¸ìœ¼ë¡œ ì‘ì„±í•˜ì„¸ìš”.**
            - ë³€í™˜ì˜ˆì‹œ:
                - ì„œìš¸ -> Seoul
                - ì¶©ë‚¨, ì¶©ì²­ë‚¨ë„ -> Chungcheongnam-do
                - ë¶€ì‚° -> Busan
        - units: str ì˜¨ë„ë‹¨ìœ„ë¥¼ ì„¤ì •í•˜ëŠ” ë¬¸ìì—´
            - metric(ê¸°ë³¸ê°’: ì„­ì”¨, ë¯¸í„°)
            - imperial(í™”ì”¨, ì•¼ë“œ)
    Return:
        - str: json í˜•ì‹ìœ¼ë¡œ ë³€í™˜ëœ í˜„ì¬ ë‚ ì”¨ ì •ë³´
    """
    url = f"https://api.openweathermap.org/data/2.5/weather?q={city_name}&appid={OPENWHEATHER_API_KEY}&units={units}"
    response = requests.get(url)  # API ìš”ì²­ ì „ì†¡
    data = response.json()        # ì‘ë‹µ(JSON)ì„ íŒŒì´ì¬ dictë¡œ ë³€í™˜
    
    weather_info = {}  # ìµœì¢… ê²°ê³¼ë¥¼ ë‹´ì„ ë”•ì…”ë„ˆë¦¬

    if response.status_code == 200:    # ì‘ë‹µ ì½”ë“œ 200ì´ë©´ ì •ìƒ ì²˜ë¦¬
        weather_description = data['weather'][0]['description']  # ë‚ ì”¨ ì„¤ëª…
        temp = data['main']['temp']                              # ì˜¨ë„
        temp_feels_like = data['main']['feels_like']             # ì²´ê° ì˜¨ë„
        humidity = data['main']['humidity']                      # ìŠµë„

        weather_info = {
            'city': city_name,
            'description': weather_description,
            'temperature': temp,
            'temperature_feels_like': temp_feels_like,
            'humidity': humidity
        }
    else:  # ì •ìƒ ì‘ë‹µì´ ì•„ë‹ˆë©´ ì‹¤íŒ¨ ì²˜ë¦¬
        weather_info = {
            'city': city_name,
            'description': 'Not Found',
            'temperature': 'Not Found',
            'temperature_feels_like': 'Not Found',
            'humidity': 'Not Found'
        }
    
    return json.dumps(weather_info)  # dictë¥¼ JSON ë¬¸ìì—´ë¡œ ë³€í™˜í•´ ë°˜í™˜
```

```python
# llmì´ ì‚¬ìš©í•  í•¨ìˆ˜(tool) ëª¨ìŒ 
tools_to_execute = {
    'get_current_weather': get_current_weather,  # í‚¤: LLMì´ í˜¸ì¶œí•  tool ì´ë¦„, ê°’: ì‹¤ì œ ì‹¤í–‰ë  í•¨ìˆ˜
}
```

```python
# Function Callingìœ¼ë¡œ ë‚ ì”¨ APIë¥¼ í˜¸ì¶œí•´ ë‹µë³€í•˜ëŠ” ëŒ€í™” ë£¨í”„
from pprint import pprint    # ë¦¬ìŠ¤íŠ¸/ë”•ì…”ë„ˆë¦¬ë¥¼ ë³´ê¸° ì¢‹ê²Œ ì¶œë ¥
from openai import OpenAI

client = OpenAI()

# ì‚¬ìš©ì ì§ˆë¬¸ì„ ë°›ì•„ tool í˜¸ì¶œ ì—¬ë¶€ë¥¼ íŒë‹¨í•˜ê³  ìµœì¢… ë‹µë³€ê¹Œì§€ ìƒì„±í•˜ëŠ” í•¨ìˆ˜
def run_conversation(user_prompt, model='gpt-4.1-mini'):
    messages = [
        # ì‹œìŠ¤í…œ í”„ë¡¬í”„íŠ¸
        {'role': 'system', 'content': 'ë‹¹ì‹ ì€ ì¹œì ˆí•œ ì±—ë´‡ì…ë‹ˆë‹¤. ì‚¬ìš©ìì˜ ìš”êµ¬ë¥¼ ë¶„ì„í•´ ì§ì ‘ ëŒ€ë‹µí•˜ê±°ë‚˜, ì£¼ì–´ì§„ í•¨ìˆ˜ë¥¼ ì´ìš©í•´ í•„ìš”í•œ ì •ë³´ë¥¼ ë¨¼ì € í™•ë³´í•œ í›„ ëŒ€ë‹µí•˜ì„¸ìš”.'},
        {'role': 'user', 'content': user_prompt}  # ìœ ì € í”„ë¡¬í”„íŠ¸
    ]
    # llmì´ ì‚¬ìš© ê°€ëŠ¥í•œ funtion(tool)ì— ëŒ€í•œ ë©”íƒ€ë°ì´í„°(ìŠ¤í‚¤ë§ˆ)
    tools = [
        {
            'type': 'function',  # toolì˜ íƒ€ì…
            'function': {
                'name': 'get_current_weather',  # LLMì´ í˜¸ì¶œí•  í•¨ìˆ˜ ì´ë¦„
                'description': tools_to_execute['get_current_weather'].__doc__,  # í•¨ìˆ˜ ì„¤ëª…(ë¬¸ì„œí™”ëœ ë¬¸ìì—´)
                'parameters': {   # í•¨ìˆ˜ì˜ ì¸ì ìŠ¤í‚¤ë§ˆ(JSON Schema)
                    'type': 'object',    # ì¸ì íƒ€ì…
                    'properties': {      # ì‚¬ìš© ê°€ëŠ¥í•œ íŒŒë¼ë¯¸í„° ì •ì˜
                        'city_name': {   # ì²« ë²ˆì§¸ íŒŒë¼ë¯¸í„°
                            'type': 'string',
                            'description': '''ë„ì‹œì´ë¦„(í•„ìˆ˜ê°’). ë°˜ë“œì‹œ ì˜ì–´ë¡œ ì‘ì„±í•˜ì„¸ìš”.
                                - ë³€í™˜ì˜ˆì‹œ:
                                    - ì„œìš¸ -> Seoul
                                    - ì¶©ë‚¨, ì¶©ì²­ë‚¨ë„ -> Chungcheongnam-do
                                    - ê²½ë‚¨, ê²½ìƒë‚¨ë„ -> Gyeongsangnam-do
                                    - ì „ë‚¨, ì „ë¼ë‚¨ë„ -> Jeollanam-do
                                    - ë¶€ì‚° -> Busan
                            ''' 
                        },  # ë„ì‹œëª… ì…ë ¥ ì•ˆë‚´(LLMì´ ì˜¬ë°”ë¥¸ ì¸ìë¥¼ ìƒì„±)
                        'units': {
                            'type': 'string',
                            'description': '''
                                ì˜¨ë„ë‹¨ìœ„ë¥¼ ì„¤ì •í•˜ëŠ” ë¬¸ìì—´
                                    - metric(ê¸°ë³¸ê°’: ì„­ì”¨, ë¯¸í„°)
                                    - imperial(í™”ì”¨, ì•¼ë“œ)
                                ''',
                            'enum': ['metric', 'imperial']  # í—ˆìš© ê°’ ì œí•œ
                        }
                    },
                    'required': ['city_name']    # í•„ìˆ˜ ì…ë ¥ê°’
                }
            }
        }
    ]

    # ì²« ë²ˆì§¸ LLM í˜¸ì¶œ (í•¨ìˆ˜ í˜¸ì¶œ í•„ìš” ì—¬ë¶€ íŒë‹¨)
    response1 = client.chat.completions.create(
        model = model,
        messages = messages,
        tools = tools    # LLMì—ê²Œ ì œê³µí•  tool ë©”íƒ€ì •ë³´
    )

    response1_message = response1.choices[0].message      # LLMì˜ ì²« ì‘ë‹µ ë©”ì‹œì§€
    response1_tool_calls = response1_message.tool_calls   # LLMì´ ìƒì„±í•œ tool í˜¸ì¶œ ëª©ë¡

    if response1_tool_calls:  # tool í˜¸ì¶œì´ ì¡´ì¬í•˜ë©´
        messages.append(response1_message)  # tool_callì´ í•¨ê»˜ ë‹´ê¸´ assistant ë©”ì‹œì§€

        for tool_call in response1_tool_calls:       # ìš”ì²­í•œ tool_call ëª©ë¡ì„ ìˆœíšŒ
            function_name = tool_call.function.name  # í˜¸ì¶œí•  í•¨ìˆ˜ ì´ë¦„
            print(f'[tool] {function_name}ì„ í˜¸ì¶œí•©ë‹ˆë‹¤!')
            function_to_execute = tools_to_execute[function_name]     # í•¨ìˆ˜ ì´ë¦„ìœ¼ë¡œ ì‹¤ì œ í•¨ìˆ˜ ê°ì²´ ì¡°íšŒ  
            function_args = json.loads(tool_call.function.arguments)  # LLMì´ ì¤€ JSON ë¬¸ìì—´ ì¸ìë¥¼ dictë¡œ íŒŒì‹±
            function_response = function_to_execute(**function_args)  # íŒŒì‹±ëœ ì¸ìë¥¼ ì–¸íŒ¨í‚¹ í›„ í•¨ìˆ˜ ì‹¤í–‰ (ì™¸ë¶€ API í˜¸ì¶œ)

            messages.append({
                'role': 'tool',  # ì—­í™œ: tool (í•¨ìˆ˜ ê²°ê³¼)
                'tool_call_id': tool_call.id,  # ì–´ë–¤ tool_callì— ëŒ€í•œ ì‘ë‹µì¸ì§€ ì—°ê²°í•˜ëŠ” ID
                'name': function_name,         # ì‹¤í–‰í•œ í•¨ìˆ˜ ì´ë¦„
                'content': function_response   # í•¨ìˆ˜ ì‹¤í–‰ ê²°ê³¼(JSON ë¬¸ìì—´)
            })
            pprint(messages)  # í˜„ì¬ê¹Œì§€ì˜ ëª¨ë“  messages í™•ì¸
        
        # tool ê²°ê³¼ë¥¼ í¬í•¨ã…ë‚˜ íˆìŠ¤í† ë¦¬ë¡œ ë‘ë²ˆì§¸ LLMì„ í˜¸ì¶œ
        response2 = client.chat.completions.create(
            model = model,
            messages = messages
        )
        return response2.choices[0].message.content  # ìµœì¢… ìì—°ì–´ ë‹µë³€ ë°˜í™˜
    
    # í˜¸ì¶œì´ ì—†ìœ¼ë©´ ì²« ì‘ë‹µì„ ê·¸ëŒ€ë¡œ ë°˜í™˜
    else:
        return response1_message.content
```
